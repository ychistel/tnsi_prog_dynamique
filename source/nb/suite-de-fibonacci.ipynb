```python
def fibo(n):
    if n==0 or n==1:
        return 1
    else:
        return fibo(n-1)+fibo(n-2)
```

```python
def fibo_mem(n):
    mem=[0]*(n+1)
    return fibo_mem_rec(n,mem)

def fibo_mem_rec(n,mem):
    if n==0 or n==1:
        mem[n]=1
        return mem[n]
    else:
        if mem[n] > 0:
            return mem[n]
        else:
            mem[n]=fibo_mem_rec(n-1,mem)+fibo_mem_rec(n-2,mem)
            return mem[n]
    
fibo_mem(274)
```
